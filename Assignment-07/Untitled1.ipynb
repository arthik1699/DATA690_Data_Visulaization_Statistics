```python
import pandas as pd
import plotly.express as px 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")

df.shape
```

```python
df.sample(5)
```

```python
my_countries = ["USA", "IND", "CHN"]

df2 = df[df["Country Code"].isin(my_countries)]

df2.shape
```

```python
df2["Country Name"].unique()
```

```python
df2.head()
```

```python
df2.tail()
```

```python
df3 = df2.query('indicator == "SP.POP.TOTL" and Year == 2021')

df3
```

```python
fig = px.bar(
    df3, 
    x="Country Name", 
    y="value", 
    title="2021 Population of USA, China, and India", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
fig = px.pie(
    df3, 
    names="Country Name", 
    values="value", 
    title="2021 Population of USA, China, and India", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

```python
fig = px.box(
    df3, 
    x ="Country Name", 
    y ="value", 
    title = "2021 Population of USA, China, and India", 
    color = "Country Name", 
    template = "plotly_dark"
    
    
)

fig.update_layout(showlegend=False)
fig.update_traces(quartilemethod="linear", jitter=0, col=1)

#fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

```python
df4 = df2.query('indicator == "SP.POP.TOTL"')

df4.head()
```

```python
fig = px.line(
    df4,
    x="Year",
    y="value", 
    title="Population of USA, China, and India over time", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.show()
```

```python
data = df_1[["Year", "value", "Region"]]

#data
df_2 = data.groupby(["Region","Year"]).sum("value")
df_2
```

```python
fig = px.bar(
    df_2, 
    x="Region", 
    y="value", 
    title="2021 Population of USA, China, and India", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```
