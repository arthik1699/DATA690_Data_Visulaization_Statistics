2.8 Calculate the minimum, maximum, and average in-state tution of the above four institutions.

```python
#Creating a smaller data frame "data" for the four universities from another smaller data frame "sdf"
data = sdf[(sdf["INSTNM"]=="Johns Hopkins University") | (sdf["INSTNM"]=="Loyola University Maryland") | 
           (sdf["INSTNM"]=="University of Maryland-College Park") | (sdf["INSTNM"]=="University of Maryland-Baltimore County")]
data
```

```python
#Finding the minimum value of the column 'TUITIONFEE_IN'
data["TUITIONFEE_IN"].min()
```

```python
#Finding the maximum value of the column 'TUITIONFEE_IN'
data["TUITIONFEE_IN"].max()
```

```python
#Finding the average value of the column 'TUITIONFEE_IN'
data["TUITIONFEE_IN"].mean()
```

3. Filtering the smaller dataset further to only keep colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia. Saving the filtered dataset to a new file named "Five_States_Colleges.csv" for future analysis.

```python
#Filtering smaller data frame to  have values from cities like Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia.
five_state_colleges = sdf[(sdf["STABBR"]=="MD")|(sdf["STABBR"]=="VA")|(sdf["STABBR"]=="DC")|(sdf["STABBR"]=="DE")|(sdf["STABBR"]=="PA")]
five_state_colleges
```

```python
#Saving the filtered data into a csv file named "Five_State_Colleges.csv"
five_state_colleges.to_csv("/content/sample_data/Five_State_Colleges.csv")
```
