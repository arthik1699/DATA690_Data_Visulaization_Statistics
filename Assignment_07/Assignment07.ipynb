```python
import pandas as pd                            # importing pandas module
import plotly.express as px                    # importing plotly.express module
```

```python
# Reading the csv file from the provided link
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")

df.shape         # finding the shape of the data frame
```

## Picking a few countries of our interest and comparing them to see how they differ in the indicator
- Comparing one year (pick the latest year) using bar chart, pie chart
- Comparing all years using line chart

#### Manipulating the data frame to get the desired set of data to plot bar and pie chart

```python
my_countries = ["ETH", "ARG", "SRB"]

df2 = df[df["Country Code"].isin(my_countries)]

df2.shape
```

```python
df2["Country Name"].unique()
```

```python
df2.head()
```

```python
df2.tail()
```

```python
df3 = df2.query('indicator == "SP.POP.TOTL" and Year == 2021')   # querying the data frame to get data of that specific indicator and year

df3
```

### Comparing one year (picking the latest year) using bar chart

```python
fig = px.bar(
    df3, 
    x="Country Name", 
    y="value", 
    title="2021 Population of USA, China, and India", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

### Comparing one year (picking the latest year) using pie chart

```python
fig = px.pie(
    df3, 
    names="Country Name", 
    values="value", 
    title="2021 Population of USA, China, and India", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

### Comparing all years using line chart

```python
df4 = df2.query('indicator == "SP.POP.TOTL"')

df4.head()
```

```python
fig = px.line(
    df4,
    x="Year",
    y="value", 
    title="Population of USA, China, and India over time", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.show()
```

### Comparing all regions to see how they differ in the indicator
- Comparing one year (picking the latest year) using bar chart, pie chart
- Comparing all years using line chart
- Note one: For population, you compare total, for the other three indicators, you compare the average.
- Note two: you need to use groupby method to get the sum or mean of a column for a group (here the group is region)

#### Manipulating the data to get desired set of data

```python
df
df_1 = df.query('indicator == "SP.POP.TOTL" and Year == 2021')
df_1
```

```python
df_2 = df_1[["Year", "value", "Region"]]

# grouping the data frame together and calculating the sum of population and resetting the index

df_3 = df_2.groupby(["Region","Year"]).sum("value").reset_index()   
df_3
```

### Comparing one year (picking the latest year) using bar chart

```python
fig = px.bar(
    df_3, 
    x="Region", 
    y="value", 
    title="Population Across the World in 2021", 
    color="Region", 
    #barmode = "group"
    template = "plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

### Comparing one year (picking the latest year) using pie chart

```python
fig = px.pie(
    df_3, 
    names="Region", 
    values="value", 
    title="Population Across the world in 2021", 
    color="Region", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

### Manipulating the data frame to get the desired set of data to plot line chart.

```python
df_4 = df.query('indicator == "SP.POP.TOTL"')
df_4
```

```python
df_5 = df_4[["Year", "value", "Region"]]
df_5
```

```python
df_6 = df_5.groupby(["Region","Year"]).sum(["value"]).reset_index()
df_6.sample(10)
```

### Comparing all years using line chart

```python
fig = px.line(
    df_6,
    x="Year",
    y="value", 
    title="Growth of Population Across Various Regions", 
    color="Region", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.show()
```

### Using all countries, and picking a year 2020, make a scatter plot with x being GDP per Capita and y being Life Expentancy. This plot will tell you the relationship between wealth and health.

#### Manipulating the data frame to get the desired set of data to plot the scatter plot

```python
df["indicator"].unique()
```

```python
my_indicators = ["SP.DYN.LE00.IN", "NY.GDP.PCAP.PP.CD"]

df_7 = df[df["indicator"].isin(my_indicators)]

df_7.shape
```

```python
df_8 = df_7.query('Year == 2020')
df_8
```

```python
df_9 = df_8[["Year",	"value",	"indicator",	"Country Code",	"Country Name"]]
          
df_9.sample(5)
```

```python
# Converting the data frame to a pivot table to make the plotting easy
df_10 = pd.pivot_table(df_9, values = 'value', index=['Country Code', 'Country Name','Year'], columns = 'indicator').reset_index()

df_10.sample(10)
```

```python
df_10.dropna(inplace=True)

df_10.sample(10)
```

### Scatter Plot showing Wealth vs Health

```python
fig = px.scatter(
    df_10,
    x="NY.GDP.PCAP.PP.CD",
    y="SP.DYN.LE00.IN", 
    title="Wealth vs Health", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)
fig.update_layout(
    xaxis_title="GDP per Capita",
    yaxis_title="Life Expectancy",)


fig.show()
```
